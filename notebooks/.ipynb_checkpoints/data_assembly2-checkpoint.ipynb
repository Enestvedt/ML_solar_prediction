```python
import pandas as pd
import os
from datetime import datetime
```

```python
data_files = os.listdir('../csv_downloads/weather')
data_files
```

```python
def load_files(data_files):
    for file in data_files:
        yield(
            pd.read_csv('../csv_downloads/weather/' + file,
            header=20,
            parse_dates= {"Time" : ["YEAR","MO","DY","HR"]},
                keep_date_col=True
            ) 
        )
```

```python
weather_data = pd.concat(load_files(data_files))
weather_data
```

```python
pd.to_datetime(weather_data['Time'])
```

```python
output_files = os.listdir('../csv_downloads/solar_output')
output_files
```

```python
def load_output(output_files):
    for file in output_files:
        yield(
            pd.read_csv('../csv_downloads/solar_output/' + file
            ) 
        )
```

```python
output_data = pd.concat(load_output(output_files))
output_data
```
