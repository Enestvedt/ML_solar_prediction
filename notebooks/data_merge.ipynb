```python
def load_output(output_files):
    for file in output_files:
        yield(
            pd.read_csv('../csv_downloads/solar_output/' + file,
            ) 
        )
```

```python
output_data = pd.concat(load_output(output_files))
output_data
```

```python
output_data['Time'] = pd.to_datetime(output_data['Time'])
```

```python
merged_data = pd.merge(weather_data, output_data, left_on='dt_iso', right_on='Time')
merged_data.sort_values(by='dt_iso')
```

```python
merged_data['month'] = merged_data['dt_iso'].dt.month
merged_data['day'] = merged_data['dt_iso'].dt.day
merged_data['hour'] = merged_data['dt_iso'].dt.hour
```

```python
cols = [
    'Time',
    'month',
    'day',
    'hour',
    'Energy (Wh)',
    'temp',
    'feels_like',
    'pressure',
    'humidity',
    'wind_speed',
    'wind_deg',
    'rain_1h',
    'rain_3h',
    'snow_1h',
    'snow_3h',
    'clouds_all',
    'weather_id',
    'weather_main',
    'weather_description'
]
```

```python
merged_data = merged_data[cols]
merged_data = merged_data.fillna(0)
merged_data
```
