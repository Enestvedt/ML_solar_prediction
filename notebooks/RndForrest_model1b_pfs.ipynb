```python
consolidated_data = pd.read_csv('../csv_downloads/merged_data.csv')
consolidated_data
```

```python
clean_consolidated_data = consolidated_data[~consolidated_data.eq(-999).any(1)]
```

```python
target = clean_consolidated_data["Energy (Wh)"]
#target_names = ["negative", "positive"]
```

```python
data = clean_consolidated_data.drop("Energy (Wh)", axis=1)
feature_names = data.columns
data.head()
```

```python
data2 = data.drop("Time", axis=1)
feature_names = data2.columns
```

```python
data = data2.drop("year", axis=1)
feature_names = data.columns
```

```python
data2 = data[[
    'month',
    # 'day',
    'hour',
    'ALLSKY_SFC_SW_DWN',
    # 'ALLSKY_KT',
    # 'SZA',
    'T2M',
    'RH2M'
    ]]
#y = clean_data["Energy (Wh)"].values.reshape(-1, 1)
#print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data2, target, random_state=42)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=150)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
